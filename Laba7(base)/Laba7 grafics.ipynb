```python
# Импротируем необходимые функции
import numpy as np # работа с массивами и линейной алгеброй
import matplotlib.pyplot as plt # для отрисовки графиков
#import matplotlib as mpl
import matplotlib.ticker as ticker
#mpl.rcParams['font.size'] = 12 # Управление стилем, в данном случаем - размером шрифта 

def line(x, a, b):
    """
    Параметрическая прямая
    """
    return a*x + b

x=np.array([-1,-2,-3])
y=np.array([0,0,-0])
y1=np.array([-4.20838,-8.18451,-11.9535])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title("Зависимость ошибки решения задачи Коши\n для уравнения " r"$y'= t^3;\quad y(0)=0$" " от шага\nв логарифмическом масштабе")

for i in range (2):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y[i+1]-y[i])/(x[i+1]-x[i])
    b=-a*x[i]+y[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(""r"$\log_{10}\,(h)$")
ax.set_ylabel("Error")

plt.show()
fig.savefig('Error for equation.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title("Зависимость ошибки решения задачи Коши\n для осциллятора " r"$y''= -y;\quad y(0)=0,\quad y'(0)=1$" " от шага\nв логарифмическом масштабе")

for i in range (2):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y1[i+1]-y1[i])/(x[i+1]-x[i])
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(""r"$\log_{10}\,(h)$")
ax.set_ylabel(""r"$\log_{10}\,(Error)$")

plt.show()
fig.savefig('Error for Oscillator.png')
```
