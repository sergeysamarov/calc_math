```python
#   
import numpy as np #      
import matplotlib.pyplot as plt #   
#import matplotlib as mpl
import matplotlib.ticker as ticker
#mpl.rcParams['font.size'] = 12 #  ,    -   

def line(x, a, b):
    """
     
    """
    return a*x + b

x=np.array([-1,-2,-3])
y=np.array([0,0,-0])
y1=np.array([-4.20838,-8.18451,-11.9535])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title("    \n   " r"$y'= t^3;\quad y(0)=0$" "  \n  ")

for i in range (2):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y[i+1]-y[i])/(x[i+1]-x[i])
    b=-a*x[i]+y[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(""r"$\log_{10}\,(h)$")
ax.set_ylabel("Error")

plt.show()
fig.savefig('Error for equation.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title("    \n   " r"$y''= -y;\quad y(0)=0,\quad y'(0)=1$" "  \n  ")

for i in range (2):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y1[i+1]-y1[i])/(x[i+1]-x[i])
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(""r"$\log_{10}\,(h)$")
ax.set_ylabel(""r"$\log_{10}\,(Error)$")

plt.show()
fig.savefig('Error for Oscillator.png')
```
