```python
# Импротируем необходимые функции
import numpy as np # работа с массивами и линейной алгеброй
import matplotlib.pyplot as plt # для отрисовки графиков
#import matplotlib as mpl
import matplotlib.ticker as ticker
#mpl.rcParams['font.size'] = 12 # Управление стилем, в данном случаем - размером шрифта 

def line(x, a, b):
    """
    Параметрическая прямая
    """
    return a*x + b

x=np.array([2,1,0,-1,-2,-3])
y1=np.array([-7.77941,-13.8592,-20.0145,-26.0518,-32.061,-38.0634])
y2=np.array([-13.5356,-21.6353,-29.797,-37.836,-45.8707,-52])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость логарифма ошибки интегрирования\n методом квадратур Гаусса c 3 узлами\n от логарифма шага')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y1[i+1]-y1[i])/(x[i+1]-x[i])
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(''r'$\log_{2}\,h$')
ax.set_ylabel(""r"$\log_{2}\,|\delta\,I_h|$")

plt.show()
fig.savefig('Gaussian quadrature with 3 nodes.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость логарифма ошибки интегрирования\n методом квадратур Гаусса с 4 узлами\n от логарифма шага')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y2[i+1]-y2[i])/(x[i+1]-x[i])
    b=-a*x[i]+y2[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(''r'$\log_{2}\,h$')
ax.set_ylabel(""r"$\log_{2}\,|\delta\,I_h|$")

plt.show()
fig.savefig('Gaussian quadrature with 4 nodes.png')

x=np.array([-1,-2,-3,-4,-5,-6,-7])
y1=np.array([-1.0769,-3.30533,-3.30533,-6.15032,-6.15032,-8.63368,-8.63368])
y2=np.array([-2.58995,-2.58995,-5.43671,-5.43671,-5.43671,-8.88752,-8.88752])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость логарифма реальной ошибки интегрирования\n методом квадратур Гаусса c 3 узлами\n от логарифма заданной точности')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y1[i+1]-y1[i])/(x[i+1]-x[i])
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(''r'$\log_{10}\,\varepsilon$')
ax.set_ylabel(""r"$\log_{10}\,|\delta\,I|$")

plt.show()
fig.savefig('Gaussian quadrature with 3 nodes with rule Runge.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость логарифма реальной ошибки интегрирования\n методом квадратур Гаусса с 4 узлами\n от логарифма заданной точности')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y2[i+1]-y2[i])/(x[i+1]-x[i])
    b=-a*x[i]+y2[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(''r'$\log_{10}\,\varepsilon$')
ax.set_ylabel(""r"$\log_{10}\,|\delta\,I|$")

plt.show()
fig.savefig('Gaussian quadrature with 4 nodes with rule Runge.png')
```
