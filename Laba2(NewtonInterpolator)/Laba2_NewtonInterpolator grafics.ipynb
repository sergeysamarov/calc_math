```python
# Импротируем необходимые функции
import numpy as np # работа с массивами и линейной алгеброй
import matplotlib.pyplot as plt # для отрисовки графиков
#import matplotlib as mpl
import matplotlib.ticker as ticker
#mpl.rcParams['font.size'] = 12 # Управление стилем, в данном случаем - размером шрифта 

def line(x, a, b):
    """
    Параметрическая прямая
    """
    return a*x + b
x=np.array([2,1,0.5,0.25,0.125,0.0625])
y1=np.array([0.213457,0.0144207,0.00133928,0.000144694,1.68266e-05,2.0291e-06])
y2=np.array([0.150654,0.00968584,0.000876828,9.37332e-05,1.09154e-05,1.31712e-06])
y3=np.array([0.0271413,0.00092402,4.29699e-05,2.32139e-06,1.34962e-07,8.13649e-09])
y4=np.array([0.0175111,0.000581822,2.71319e-05,1.4675e-06,8.53627e-08,5.14759e-09])
y5=np.array([0.00305649,5.26572e-05,1.22995e-06,3.32864e-08,9.68434e-10,2.9204e-11])
y6=np.array([0.00167901,2.89663e-05,6.77009e-06,1.83271e-08,5.33277e-10,1.60827e-11])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Ньютона\n от длины отрезка для равномерного расположения узлов, N=3')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y1[i+1]-y1[i])/(x[i+1]-x[i])
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')

plt.show()
fig.savefig('Uniform points N=3.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Ньютона\n от длины отрезка для чебышевского расположения узлов, N=3')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y2[i+1]-y2[i])/(x[i+1]-x[i])
    b=-a*x[i]+y2[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Chebishev points N=3.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Ньютона\n от длины отрезка для равномерного расположения узлов, N=4')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y3[i+1]-y3[i])/(x[i+1]-x[i])
    b=-a*x[i]+y3[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Uniform points N=4.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Ньютона\n от длины отрезка для чебышевского расположения узлов, N=4')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y4[i+1]-y4[i])/(x[i+1]-x[i])
    b=-a*x[i]+y4[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')

plt.show()
fig.savefig('Chebishev points N=4.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Ньютона\n от длины отрезка для равномерного расположения узлов, N=5')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y5[i+1]-y5[i])/(x[i+1]-x[i])
    b=-a*x[i]+y5[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Uniform points N=5.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Ньютона\n от длины отрезка для чебышевского расположения узлов, N=5')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y6[i+1]-y6[i])/(x[i+1]-x[i])
    b=-a*x[i]+y6[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Chebishev points N=5.png')

```
