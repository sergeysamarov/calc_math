```python
# Импротируем необходимые функции
import numpy as np # работа с массивами и линейной алгеброй
import matplotlib.pyplot as plt # для отрисовки графиков
#import matplotlib as mpl
import matplotlib.ticker as ticker
#mpl.rcParams['font.size'] = 12 # Управление стилем, в данном случаем - размером шрифта 

def line(x, a, b):
    """
    Параметрическая прямая
    """
    return a*x + b
x=np.array([2,1,0.5,0.25,0.125,0.0625])
y1=np.array([0.000631092,5.57733e-06,6.60057e-08,8.99189e-10,1.31248e-11,1.98286e-13])
y2=np.array([0.000271486,2.33587e-06,2.77526e-08,3.78762e-10,5.53291e-12,8.37108e-14])
y3=np.array([2.93672e-06,6.54056e-09,1.93903e-11,6.61693e-14,2.22045e-16,2.22045e-16])
y4=np.array([1.16778e-06,2.61184e-09,7.75535e-12,2.66454e-14,2.22045e-16,2.22045e-16])
y5=np.array([1.06178e-08,5.96323e-12,4.66294e-15,2.22045e-16,2.22045e-16,2.22045e-16])
y6=np.array([3.21647e-09,1.80789e-12,1.55431e-15,2.22045e-16,2.22045e-16,2.22045e-16])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Эрмита\n от длины отрезка для равномерного расположения узлов, N=3')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y1[i+1]-y1[i])/(x[i+1]-x[i])
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')

plt.show()
fig.savefig('Uniform points N=3.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Эрмита\n от длины отрезка для чебышевского расположения узлов, N=3')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y2[i+1]-y2[i])/(x[i+1]-x[i])
    b=-a*x[i]+y2[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Chebishev points N=3.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Эрмита\n от длины отрезка для равномерного расположения узлов, N=4')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y3[i+1]-y3[i])/(x[i+1]-x[i])
    b=-a*x[i]+y3[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Uniform points N=4.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Эрмита\n от длины отрезка для чебышевского расположения узлов, N=4')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y4[i+1]-y4[i])/(x[i+1]-x[i])
    b=-a*x[i]+y4[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')

plt.show()
fig.savefig('Chebishev points N=4.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Эрмита\n от длины отрезка для равномерного расположения узлов, N=5')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y5[i+1]-y5[i])/(x[i+1]-x[i])
    b=-a*x[i]+y5[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Uniform points N=5.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость ошибки интерполяции полиномами Эрмита\n от длины отрезка для чебышевского расположения узлов, N=5')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y6[i+1]-y6[i])/(x[i+1]-x[i])
    b=-a*x[i]+y6[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel('Длина отрезка')
ax.set_ylabel('Ошибка интерполяции')


plt.show()
fig.savefig('Chebishev points N=5.png')

```
