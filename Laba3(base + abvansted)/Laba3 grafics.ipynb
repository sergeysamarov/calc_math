```python
# Импротируем необходимые функции
import numpy as np # работа с массивами и линейной алгеброй
import matplotlib.pyplot as plt # для отрисовки графиков
#import matplotlib as mpl
import matplotlib.ticker as ticker
#mpl.rcParams['font.size'] = 12 # Управление стилем, в данном случаем - размером шрифта 

def line(x, a, b):
    """
    Параметрическая прямая
    """
    return a*x + b

x=np.array([2.32193,3.32193,4.32193,5.32193,6.32193,7.32193])
y1=np.array([11.605,9.92902,8.03656,6.06444,4.07282,2.04846])
y2=np.array([11.8925,10.0187,8.11253,6.13756,4.14503,2.12232])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость логарифма ошибки интерполяции\n от логарифма количества узлов')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=y1[i+1]-y1[i]
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(''r'$\log_{2}\,N$')
ax.set_ylabel(""r"$\log_{2}\,\sup_{[0,10]}\;|I\,(x)-e^x|$")

plt.show()
fig.savefig('Natural spline.png')

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title('Зависимость логарифма ошибки интерполяции\n от логарифма количества узлов')

for i in range (5):
    z = np.linspace(x[i], x[i+1], 1000)
    a=y2[i+1]-y2[i]
    b=-a*x[i]+y2[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(''r'$\log_{2}\,N$')
ax.set_ylabel(""r"$\log_{2}\,\sup_{[0,10]}\;|I\,(x)-e^x|$")

plt.show()
fig.savefig('Spline with real means second derivatives.png')
```
