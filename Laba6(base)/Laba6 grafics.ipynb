```python
#   
import numpy as np #      
import matplotlib.pyplot as plt #   
#import matplotlib as mpl
import matplotlib.ticker as ticker
#mpl.rcParams['font.size'] = 12 #  ,    -   

def line(x, a, b):
    """
     
    """
    return a*x + b

x=np.array([0,-1,-2,-3,-4,-5,-6,-7])
y=np.array([-45,-49,-53,-57,-61,-65,-69,-73])
#x1=np.array([0,1,2])
y1=np.array([-4.93459,-10.6182,-15.4169,-19.7904,-24.6345,-29.5972,-34.588,-39.5857])
#x2=np.array([0,1,2,3])
#y2=np.array([-0.322115,-1.15111,-2.86484,-6.28138])
#x3=np.array([0,1,2,3,4,5])
#y3=np.array([-0.0969558,-0.29881,-1.17045,-2.76724,-5.93808,-12.2792])

fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title("    \n   " r"$y'= t^3;\quad y(0)=0$" "  \n  ")

for i in range (7):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y[i+1]-y[i])/(x[i+1]-x[i])
    b=-a*x[i]+y[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(""r"$\log_{2}\,(h)$")
ax.set_ylabel(""r"$\log_{2}\,(Err)$")

plt.show()
fig.savefig('Error for equation.png')

#fig=plt.figure(figsize=(7,7))
ax = plt.subplot(111)

ax.set_title("    \n   " r"$y''= -y;\quad y(0)=0,\quad y'(0)=1$" "  \n  ")

for i in range (7):
    z = np.linspace(x[i], x[i+1], 1000)
    a=(y1[i+1]-y1[i])/(x[i+1]-x[i])
    b=-a*x[i]+y1[i]
    ax.plot(z, line(z, a, b), 'b')

ax.grid(which = "major", linewidth = 1)
ax.grid(which = "minor", linewidth = 0.2)
ax.minorticks_on()
ax.set_xlabel(""r"$\log_{2}\,(h)$")
ax.set_ylabel(""r"$\log_{2}\,(Err)$")

plt.show()
fig.savefig('Error for Oscillator.png')
```
